# Pokemon Stats Analysis 

In this notebook I will work with a database containing stats and other useful data, like types and legendary status, of 721 Pokemon (up to Generation VI). My goal will be to explore the information of all Pokemon and find trends, in order to finally get a list of the top 10 best non-legendary Pokemon and the top 5 legendary Pokemon.

### Importing the data

```python
# We import all the necessary libraries.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# We set the style of the Seaborn graphs.
sns.set_style('darkgrid')

# We read the flat file containing the data and create a dataframe based on it.
pokemon = pd.read_csv('Pokemon.csv')

# We print the head of the dataframe to verify that the data was imported properly.
pokemon.head()
```

```python
# The data was imported correctly, so let's get a sample of 10 Pokemon from the dataset to see what information is available.
pokemon.sample(10)
```

### Exploratory Data Analysis

```python
# Let's get some general info of the data frame.
pokemon.info()
```

There are 800 unique entries or Pokemon, with 13 variables categorizing them. We got the Name of the Pokemon, its primary and secondary type when it applies, the combat stats of the Pokemon (Total Stats, HP, Attack, Defense, Special Attack, Special Defense, and Speed), what generation it is from and whether it is legendary or not.

```python
# There doesn't seem to be any missing values, but let's change the nulls from the Type 2 column into "None".
pokemon['Type 2'].fillna('None', inplace=True)
# Let's verify that the change was applied.
pokemon['Type 2'].isnull().sum()
```

```python
# Let's check for duplicates in the dataset.
pokemon.duplicated().sum()
```

```python
# Descriptive statistic summary of the stats of the Pokemon.
pokemon.describe()
```

```python
# Let's visualize the distribution of the data.

# Distribution of the total stat sum of all Pokemon.
sns.displot(pokemon['Total'], bins=20, color = 'orange')
plt.show()
```

Most Pokemon have a stat Total of about 490-500.

```python
# Number of Pokemon by Generation
plt.figure(figsize=(10, 6))
sns.countplot(x='Generation', data=pokemon)
plt.xlabel('Generation')
plt.ylabel('Number of Pokemon')
plt.title('Pokemon by Generation')
plt.show()
```

Generation 1 is the generation most represented in the dataset, followed closely by Gen V and Gen III. The least represented Generation is the 6th.

```python
# Counts of legendary vs non-legendary pokemon.
plt.figure(figsize=(10, 6))
sns.countplot(x='Legendary', data=pokemon, order=[True, False])
plt.xlabel('Legendary')
plt.ylabel('Number of Pokemon')
plt.title('Pokemon by Legendary status')
plt.show()
```

```python
# Distribution of Attack and Defense, relationship between both.
sns.jointplot(data=pokemon, x='Attack', y='Defense',kind='reg')
plt.show()
```

There seems to be a positive relationship between Attack and Defense, when one goes up, the other one increases as well.

```python
# Attack and Defense by primary Type.
plt.figure(figsize=(10,10))
sns.scatterplot(data=pokemon, x='Attack', y='Defense',hue='Type 1')
plt.show()
```

Most Pokemon have an Attack of about 60 to 70, while most Pokemon have a Defense of around 55 to 65.
Looking at the data, it would seem that Normal and Bug Pokemon are often the worst performing in these two attributes, while Dragon, Rock, Ghost and Steel seem to be especially strong.

```python
# Distribution of Special Attack and Special Defense, relationship between both.
sns.jointplot(data=pokemon, x='Sp. Atk', y='Sp. Def', kind='reg')
plt.show()
```

The relationship between Sp. Atk and Sp. Def is almost identical to that of regular Attack and Defense.

```python
# Special Attack and Special Defense by Primary Type
plt.figure(figsize=(10,10))
sns.scatterplot(data=pokemon, x='Sp. Atk', y='Sp. Def', hue='Type 1')
plt.show()
```

Special Attack seems to vary a lot from type to type, with a lot of Normal Pokemon having Sp. Atk below 50, Water Pokemon being evenly distributed from about 55 to 85 Sp. Atk. Special Defense is less diverse, with most Pokemon having this stat at about 60 to 70. Once again, Normal and Bug Pokemon seem to be very weak in these attributes, and there seem to be less strong outliers in the data.

```python
# Let's see the distribution of the Speed stat.
sns.displot(pokemon['Speed'], bins = 20)
plt.title('Speed stat distribution')
plt.show()
```

Most Pokemon are moderately slow, with most having a speed of about 55 to 60. Speed is king in competitive Pokemon, so those few over 100 are going to be great performers in a competitive environment.

```python
# Distribution of the HP stat.
sns.displot(pokemon['HP'], color = 'red', bins=20)
plt.title('HP stat distribution')
plt.show()
```

Most Pokemon have base HP of about 60 to 70.

```python
# In order to get the correlation between stats, a new dataframe containing only the 6 stats will be created.
corr_df = pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Stat correlation map
plt.rcParams['figure.figsize']=(15,9)
heat_map = sns.heatmap(corr_df.corr(), annot = True, linewidths=1, cmap='Blues')
heat_map.set_title(label='Stat Correlation Map', fontsize=25)
heat_map;
```

The closest correlations among stats are between Defense and Special Defense and Special Defense and Special Attack. Special Attack and Speed seem to have a relatively high correlation as well.

Speed and Defense are barely correlated at all, and the relationship between Speed and HP is negligible as well.

### Defining competitiveness of a Pokemon

While Pokemon have 6 stats, when it comes to high level competitive play some stats are more useful than others. This is why a simple sum of all stats, like the one in the dataset, would be useless in order to find the best Pokemon. 

While there are special cases and strategies where this does not apply, Speed is usually the most important stat of a Pokemon. Making your move first can be the difference between life and death.
